##Downlad from PubChem BioAssay with RESTful API
Example: http://pubchem.ncbi.nlm.nih.gov/rest/pug/assay/aid/504526/CSV 

```python
import json
import urllib2
from pprint import pprint
```

```python
my_url = "http://pubchem.ncbi.nlm.nih.gov/rest/pug/assay/aid/504526/JSON"

my_json = json.load(urllib2.urlopen(my_url))

pprint(my_json)
```

```python
my_url = "http://pubchem.ncbi.nlm.nih.gov/rest/pug/assay/aid/504526/CSV"
for line in urllib2.urlopen(my_url).readline():
    print line
```
