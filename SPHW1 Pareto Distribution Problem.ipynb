```python
u = np.random.uniform(0, 1, size=ss)
plt.hist(u)
plt.title("Numpy Sampling from Uniform(0, 1)")
plt.show()
```

```python
a = 15.0
p = np.random.pareto(a, size=ss)
plt.hist(p)
plt.show()
```

```python
p_simulated = (1-u)**(-1/a)-1
plt.hist(p_simulated)
plt.show()
```
