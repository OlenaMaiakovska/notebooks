```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(seed=126)
samples = 20000
```

##Part A: Build Simulators from Uniform(0, 1)

```python
plt.hist(np.random.uniform(0, 1, size=samples))
plt.title("Numpy Sampling from Uniform(0, 1)")
plt.show()
```

```python
a, b = 0, np.pi
T = (b - a) * np.random.uniform(0, 1, size=samples) + a

plt.hist(T)
plt.title(r"Calculated: $T = 2\pi * u$ ~ Uniform(0, $2\pi$)")
plt.show()
```

```python
U = np.negative(np.log(1 - np.random.uniform(0, 1, size=samples)))

plt.hist(U)
plt.title(r"Calculated: $U = -\ln(1-u)$ ~ Exponential(1) ")
plt.show()
```

```python
X = np.sqrt(2 * U) * np.cos(T)

plt.hist(X)
plt.title("Calculated: $X = \sqrt{2U}\cos(T)$")
plt.show()
```

```python
Y = np.sqrt(2 * U) * np.sin(T)

plt.hist(Y)
plt.title("Calculated: $Y = \sqrt{2U}\sin(T)$")
plt.show()
```
