```python
import networkx as nx

import numpy as np
from numpy.random import uniform

import itertools as itt
from itertools import combinations, product, accumulate

import pandas as pd
import scipy as scp
from scipy import stats

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (16.0, 9.0)
np.random.seed(seed=127)
```

```python
class MGParams:
    def __init__(self, sizes, p, colors=None):
        self.sizes = sizes
        self.p = np.array(p)
        self.colors = colors
        self.n = len(sizes)
        
        if not len(sizes) == len(p) == len(colors):
            raise 'invalid input'

class MG:
    def __init__(self, params):
        self.params = params
        self.node_groups = {}
        self.s_edge_sets = {}
        self.c_edge_sets = {}
        self.pos = None
        self.graph = nx.Graph()

    def nrange(self):
        return range(self.n)
        
    def generate(self):
        prev = 0
        for group, acc in enumerate(accumulate(self.params.sizes)):
            self.node_groups[group] = list(range(prev, acc))
            prev = acc

        for i, node_group in self.node_groups.items():
            self.s_edge_sets[i] = [x for x in combinations(node_group, 2) if uniform() < self.params.p[i, i]]

        for c1, c2 in combinations(range(self.params.n), 2):
            self.c_edge_sets[c1, c2] = [x for x in product(self.node_groups[c1], self.node_groups[c2]) if uniform() < self.params.p[c1, c2]]

    def generate_nx(self):
        for i, node_group in self.node_groups.items():
            for node in node_group:
                self.graph.add_node(node, group=i)

        for s_edge_set in self.s_edge_sets.values():
            for edge in s_edge_set:
                self.graph.add_edge(*edge)

        for c_edge_set in self.c_edge_sets.values():
            for edge in c_edge_set:
                self.graph.add_edge(*edge)

    def position(self):
        self.pos = nx.circular_layout(self.graph)
                
    def draw(self, title=None):
        for i, node_group in self.node_groups.items():
            nx.draw_networkx_nodes(self.graph, 
                                   self.pos, 
                                   nodelist=node_group, 
                                   node_color=self.params.colors[i])

        for i, s_edge_set in self.s_edge_sets.items():
            nx.draw_networkx_edges(self.graph, 
                                   self.pos, 
                                   edgelist=s_edge_set, 
                                   edge_color=self.params.colors[i],  
                                   width=5, 
                                   alpha=0.5)

        for c_edge_set in self.c_edge_sets.values():
            nx.draw_networkx_edges(self.graph, 
                                   self.pos, 
                                   edgelist=c_edge_set, 
                                   edge_color='grey',  
                                   width=5, 
                                   alpha=0.5) 
        if title:
            plt.title(title)
        plt.axis('off')
        plt.show()
    
    def do(self, title=None):
        self.generate()
        self.generate_nx()
        self.position()
        self.draw(title)
        
t1 = MGParams(sizes=[4,4], p=[[0.3, 0.3], [0.3, 0.3]], colors=["red", "blue"])
MG(t1).do("Example 4 Graph")
```
