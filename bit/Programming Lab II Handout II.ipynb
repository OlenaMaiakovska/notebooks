```python
def is_triangle(a, b, c):
    x = sorted((a, b, c))
    return x[2] <= x[1] + x[0]

assert is_triangle(3,4,5)
assert is_triangle(3,5,4)
assert is_triangle(5,4,3)
assert not is_triangle(1,2,12)
assert not is_triangle(12,1,2)
assert not is_triangle(1,12,2)
```

# Exercise 2

```python
def gcd(a, b):
    if 0 == b:
        return a
    elif b > a:
        return gcd(b, a)
    return gcd(b, a % b)

assert 5 == gcd(20, 5)
assert 5 == gcd(5, 20)
assert 5 == gcd(5, 0)
assert 5 == gcd(0, 5)
assert 1 == gcd(7, 13)
```

# Exercise 3

```python
import math

c = 2 * math.sqrt(2) / 9801

def estimate_pi_helper(it):
    pi_inv = 0
    for k in range(it):
        f = math.factorial(4 * k)
        f *= (1103 + 26390 * k)
        f /= math.factorial(k) ** 4
        f /= 396 ** (4 * k)
        pi_inv += f
    return 1 / (c * pi_inv)

def estimate_pi(precision=1e-15):
    i = 1
    pi = estimate_pi_helper(i)
    while (math.fabs(pi - math.pi) > precision):
        i += 1
        pi = estimate_pi_helper(i)
    return pi

assert estimate_pi() - math.pi < 1e-15
```

# Exercise 4

```python
def digitize(n):
    x = []
    y = n
    while(10 <= y):
        x.append(y % 10)
        y //= 10
    x.append(y)
    return list(reversed(x))

assert [1, 2, 3] == digitize(123)
assert [1, 2] == digitize(12)
assert [1, 0, 0] == digitize(100)
assert [1, 0, 1] == digitize(101)
assert [1, 0] == digitize(10)
assert [1] == digitize(1)
assert [0] == digitize(0)

def isHappy(n):
    if n < 0:
        return False
    elif n == 1:
        return True
    elif n == 4:
        return False
    nn = sum(digit ** 2 for digit in digitize(n))
    return isHappy(nn)

for n in [1, 7, 10, 13, 19, 23]:
    assert isHappy(n)

for n in [2, 3, 4, 5, 6]:
    assert not isHappy(n)
```

## Exercise 5 A

```python
import re
import functools

def avoids1(word, forbidden):
    for a in word:
        for b in forbidden:
            if a == b:
                return False
    return True

def avoids2(word, forbidden):
    return all(char not in word for char in forbidden)

def avoids3(word, forbidden):
    return not re.search("[{}]".format(forbidden), word)

for fn in (avoids1, avoids2, avoids3):
    assert fn("hello", "xyz"), fn
    assert not fn("hello", "habc"), fn
    assert not fn("hello", "abhc"), fn
```

## Exercise 5B

```python
def uses_only(word, allowed):
    for ch in word:
        if ch not in allowed:
            return False
    return True

def uses_only2(word, allowed):
    return all(ch in allowed for ch in word)

for fn in (uses_only, uses_only2):
    assert fn("hello", "oleh"), fn
    assert not fn("hello", "hlox"), fn
    assert fn("hello", "elxoh"), fn

def filter_words(words, allowed):
    return list(filter(lambda word: uses_only(word, allowed), words))

filter_words(words[:1000], "cab")
```

## Exercise 5C

```python
def uses_all(word, required):
    return set(word) >= set(required)

def uses_all2(word, required):
    return uses_only(required, word)

for fn in (uses_all, uses_all2):
    assert fn("hello", "helo"), fn
    assert fn("heloohlel", "helo"), fn
    assert fn("abcd", "abc"), fn
    assert not fn("heloohlel", "hexlo"), fn

ex_5c_a = [word for word in words if uses_all(word, "aeiou")]
ex_5c_b = [word for word in words if uses_all(word, "aeiouy")]
    
print("# words with aeiou {}\n# words with aeiouy {}".format(len(ex_5c_a), len(ex_5c_b)))
```

```python
def is_abecedarian(word):
    return list(word) == list(sorted(word))

assert is_abecedarian('abcess')
assert not is_abecedarian('axzyab')
```

```python
def triple_double(w):
    for i in range(len(w) - 5):
        if w[i] == w[i+1] and w[i+2] == w[i+3] and w[i+4] == w[i+5]:
            return True
    return False

assert triple_double("bookkeeper")
assert not triple_double("housekeeper")
```
