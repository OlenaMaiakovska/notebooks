## Action I

Compute a 150 x 150 Euclidean distance matrix between flowers. You can use R- function dist for that purpose, if you want. (4 points + 1 bonus)

```r
d = as.matrix(dist(iris[1:4]))

head(d)
```

## Action II

Given that distance matrix calculate for each flower its so-called nearest neighbor (i.e. the one, which is most similar to it). Print the corresponding table on the screen and write it into a .csv file. (4 points + 1 bonus)

```r
nearest_neighbor = vector(length=n)
nearest_neighbor_species = vector(length=n)

for (i in 1:n) {
    d[i, i] = NA
    nearest_neighbor[i] = which.min(d[i,])
    nearest_neighbor_species[i] = iris$Species[nearest_neighbor[i]]
}

nearest_neighbor_species = factor(nearest_neighbor_species, labels=unique(iris$Species))

df = data.frame(iris$Species, nearest_neighbor, nearest_neighbor_species)
names(df) = c("Species", "NN", "NN.Species")

write.table(df, file="alb_assignment1_action2.csv")

head(df)
```
